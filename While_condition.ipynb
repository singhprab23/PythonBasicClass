# write while loop to print number from 1 to 5


```python
num=1
while num<=5:
    print(num)
    num+=1
```

```python
num=1
while num<=10:
  if num==4:
    print(num)
  if num%2==0:
    print("Even:",num)
  if num%2!=0:
    print("Odd:",num)
  num+=1
```

```python
x = 1
while x <= 20:
    if x % 3 == 0 and x % 5 == 0:
        print("FizzBuzz",x)
    elif x % 3 == 0:
        print("Fizz",x)
    elif x % 5 == 0:
        print("Buzz",x)
    else:
        print(x)
    x += 1
```

```python
#end and sep parameters

num=1
while num<=5:
  print(num,end=",")
  num+=1
```

```python
num=1
while num<=5:
  if num==5:
    print(num)
  else:
    print(num,end=",")
  num+=1
```
