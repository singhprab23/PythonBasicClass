```python
# positional arguments: position matters

def family(father, mother, sibling):
  print("Father name is:", father)
  print("Mother name is:", mother)
  print("Sibling name is:", sibling)
```

```python
family("Prabhjot Sandhu","Ramanpreet Sandhu ","Yuvraj Sandhu")
```

```python
family("Ramanpreet Sandhu","Prabhjot Sandhu","Yuvraj Sandhu","C")
#
```

```python
# keyword Argument

def family(father, mother, sibling):
  print("Father name is:", father)
  print("Mother name is:", mother)
  print("Sibling name is:", sibling)
```

```python
def do_something(n):
    total = 0
    i = 1
    while i <= n:
        total += i
        i += 1
    return total

print(do_something(3))
```

```python
# scope of variables
num=5 # global variable
def func():
  num=10 # variable local to the function
  print("inside func")
  print(num)

func()
print("outside func")
print(num)
```

```python
# scope of variables
num=5 # global variable
def func():
  #num=10 # variable local to the function
  global num
  num+=2
  print("inside func")
  print(num)

func()
print("outside func")
print(num)
```
